```python
#Encoding the data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df=df.apply(le.fit_transform)
df.head()
```

```python
#4.Feature Selection

# # Dividing dependent and independent variables
X = df.drop("Sleep Disorder", axis=1)
y = df["Sleep Disorder"]
```

```python
# Sampling
from imblearn.over_sampling import SMOTE 
from collections import Counter

sampler = SMOTE()
X, y = sampler.fit_resample(X, y)

print('SMOTE {}'.format(Counter(y)))
```

```python
#5.ModelBuilding

#Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression  

classifier= LogisticRegression(random_state=0)  
lr_model=classifier.fit(X_train, y_train)
y_pred=lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
x_axis_labels = ['Not S Disorder','Sleep Disorder']
y_axis_labels = ['Not S Disorder','Sleep Disorder']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

```python
#Dicision Tree
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
dt_model = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = dt_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
x_axis_labels = ['Not S Disorder','Sleep Disorder']
y_axis_labels = ['Not S Disorder','Sleep Disorder']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=25)

# Train the model using the training sets
knn_model=clf.fit(X_train,y_train)

y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
x_axis_labels = ['Not S Disorder','Sleep Disorder']
y_axis_labels = ['Not S Disorder','Sleep Disorder']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100,random_state=0)
rf_model=clf.fit(X_train, y_train)
y_pred=rf_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
x_axis_labels = ['Not S Disorder','Sleep Disorder']
y_axis_labels = ['Not S Disorder','Sleep Disorder']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```
